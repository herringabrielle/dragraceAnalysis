SPRING 2019

This will grab datapoints by timestamp from the server and save them to my local device. It will name them with the station number, time stamp, Spring race, and component. It saves a time window 10 seconds before to 70 seconds after the timestamp.  The saved file is named the same time stamp as before.  It is not named for the 10 s before capture window.  

```python
# SPRING

window = 60 # Window of time to save the data, in seconds
count = 0

with open('timestamp_Sp_ALL.txt', 'r') as file_in:
 
  for line in file_in:
        try:
            line = line.rstrip()
            start = parse(line)
            start = start+timedelta(seconds=-10)
            for sta in ['FL','NL','NR','FR']:
                stx = dragrace.getMseed(start, window, staList=[sta,], chanList= ['HNX'] )
                sty = dragrace.getMseed(start, window, staList=[sta,], chanList= ['HNY'] )
                stz = dragrace.getMseed(start, window, staList=[sta,], chanList= ['HNZ'] )
                
                # Filepath dependent 
                
                stx.write("./springClean/Original/SP_" + str(sta) + "_"  + line + "_HNX.mseed", format='MSEED') 
                sty.write("./springClean/Original/SP_" + str(sta) + "_"  + line + "_HNY.mseed", format='MSEED') 
                stz.write("./springClean/Original/SP_" + str(sta) + "_"  + line + "_HNZ.mseed", format='MSEED')
                #print("{}  {}  {}  {}".format(sta, stx[0][0], sty[0][0], stz[0][0]))
                count +=1
       

        except Exception as e:
            print(e, "error")
        #break # Use this to test the first case before you run the entire script, it takes a long time
print(count,'number of files')
```
